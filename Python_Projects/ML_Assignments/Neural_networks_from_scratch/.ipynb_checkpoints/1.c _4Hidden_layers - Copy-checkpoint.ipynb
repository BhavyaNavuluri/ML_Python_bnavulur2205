```python
import numpy as np
import pandas as pd 
import csv 
from sklearn.preprocessing import OneHotEncoder 
from sklearn.model_selection import train_test_split
from scipy.io import loadmat
import warnings
warnings.simplefilter("ignore")
```

```python
#  loading the data set 
file = pd.read_csv('BSOM_DataSet_for_HW3.csv')


# to check whether there are any empty values or not 
df=file[['all_mcqs_avg_n20', 'all_NBME_avg_n4', 'CBSE_01',  'CBSE_02', 'LEVEL']]


df['LEVEL'].fillna(value='A', inplace=True) 

df[['CBSE_02']].fillna(method ='bfill')
#.fillna(52, inplace=True)
df.fillna(df['CBSE_02'].mean(), inplace=True)

df.isnull().any()
```

```python
LEVEL = {'A': 0,'B': 1,'C':2,'D':3}

df.LEVEL = [LEVEL[item] for item in df.LEVEL] 
print(df) 
```

```python
# Transforming y to binary number using one-hot- encoding
onhtencder = OneHotEncoder(categorical_features = [0]) 
data = onhtencder.fit_transform(df[['LEVEL']]).toarray() 
#data
```

```python

X= df.iloc[:,:]

X_norm = (X.iloc[:,:-1] - X.iloc[:,:-1].mean()) / (X.iloc[:,:-1].max() - X.iloc[:,:-1].min())



X_modified = pd.concat([X_norm, X.iloc[:,-1]],axis=1)

target=df.iloc[:,-1]

X_train, X_test, y_train, y_test = train_test_split(X_modified, data, test_size=0.20,random_state=14)
```

```python

Train_ip=np.array(X_train.iloc[:,:-1])
Train_otpts=np.array(y_train)
```

```python
# Initializing weights randomly


np.random.seed(3)

wt1 =  np.random.random((1,5))  


wt2 =  np.random.random((2,2))   

wt3 =  np.random.random((2,3))  

wt4 = np.random.random((2,3))  

wt5 =  np.random.random((4,3))   


ttl_weits = np.hstack((wt1.ravel(order='F'), wt2.ravel(order='F'),wt3.ravel(order='F'),wt4.ravel(order='F'),wt5.ravel(order='F')))    

# layers configuration 

ip_lyr = 4
hd_lyr1 = 1
hd_lyr2 = 2
hd_lyr3 = 2
hd_lyr4 = 2
no_labls = 4

print('wt1',wt1)
print('wt2',wt2)
print('wt3',wt3)
print('wt4',wt4)
print('wt5',wt5)
```

```python
def Weights(ttl_weits, ip_lyr, hd_lyr1,hd_lyr2,hd_lyr3,hd_lyr4, no_labls):

    
    w1 = np.reshape(ttl_weits[:hd_lyr1*(ip_lyr+1)], (hd_lyr1, ip_lyr+1), 'F')
    w2 = np.reshape(ttl_weits[(ip_lyr+1):((ip_lyr+1)+hd_lyr2*(hd_lyr1+1))], (hd_lyr2, hd_lyr1+1), 'F')
    w3 = np.reshape(ttl_weits[((ip_lyr+1)+hd_lyr2*(hd_lyr1+1)):(((ip_lyr+1)+hd_lyr2*(hd_lyr1+1))+hd_lyr3*(hd_lyr2+1))],(hd_lyr3, hd_lyr2+1), 'F')
    w4 = np.reshape(ttl_weits[((ip_lyr+1)+hd_lyr2*(hd_lyr1+1)+hd_lyr3*(hd_lyr2+1)):(((ip_lyr+1)+hd_lyr2*(hd_lyr1+1)+hd_lyr3*(hd_lyr2+1))+hd_lyr4*(hd_lyr3+1))], (hd_lyr4, hd_lyr3+1), 'F')
    w5 = np.reshape(ttl_weits[((ip_lyr+1)+hd_lyr2*(hd_lyr1+1)+hd_lyr3*(hd_lyr2+1)+hd_lyr4*(hd_lyr3+1)):], (no_labls, hd_lyr4+1), 'F')
    
   
    return w1,w2,w3,w4,w5

    
```

```python
def Cost(h,y):
    cost = np.sum(np.multiply(y, np.log(h))+np.multiply(1-y, np.log(1-h)))
    return cost
```

```python


def SIG(z):
    return 1/(1+np.exp(-z))



def SIGGrad(z):
    return np.multiply(SIG(z), 1-SIG(z))
```

```python
def Feedfrwd(ttl_weits, ip_lyr, hd_lyr1,hd_lyr2,hd_lyr3,hd_lyr4, no_labls, Train_ip, Train_otpts):
    
    wt1,wt2,wt3,wt4,wt5 = Weights(ttl_weits, ip_lyr, hd_lyr1,hd_lyr2,hd_lyr3,hd_lyr4, no_labls)
    
    l = len(Train_otpts)
   
    ones = np.ones((l,1))
    
    a1 = np.hstack((ones, Train_ip))
    
    a2 = SIG(a1 @ wt1.T)
    
    a2 = np.hstack((ones, a2))
    
    a3 = SIG(a2 @ wt2.T)
    
    a3 = np.hstack((ones, a3))
    
    a4 = SIG(a3 @ wt3.T)
    
    a4 = np.hstack((ones, a4))
    
    a5 = SIG(a4 @ wt4.T)
    
    a5 = np.hstack((ones, a5))
    
    h = SIG(a5 @ wt5.T)
    
    c=Cost(h,Train_otpts)
    
   
    
    return c / (-l)
    
```

```python
cf_er = Feedfrwd(ttl_weits, ip_lyr, hd_lyr1,hd_lyr2,hd_lyr3,hd_lyr4, no_labls, Train_ip,Train_otpts)
cf_er
```

```python
def NN(ttl_weits, ip_lyr, hd_lyr1,hd_lyr2,hd_lyr3, hd_lyr4,no_labls, X, y):
    
    
    initt_wt1,initt_wt2,initt_wt3,initt_wt4,initt_wt5 = Weights(ttl_weits, ip_lyr, hd_lyr1,hd_lyr2,hd_lyr3,hd_lyr4, no_labls)

    l = len(y)

    ones = np.ones((l,1))
    a1 = np.hstack((ones, X))
    #print('a1',a1)
    print('a1',a1.shape)
    z2 = a1 @ initt_wt1.T
    #print('z2',z2)
    print('z2',z2.shape)
    a2 = np.hstack((ones, SIG(z2)))
    #print('a2',a2)
    print('a2',a2.shape)
    z3 = a2 @ initt_wt2.T
    #print('z3',z3)
    print('z3',z3.shape)
    a3 = np.hstack((ones, SIG(z3)))
    #print('a3',a3)
    print('a3',a3.shape)
    #print('',.shape)print('',)
    z4 = a3 @ initt_wt3.T
    
    a4 = np.hstack((ones, SIG(z4)))
    
    z5 = a4 @ initt_wt4.T
    
    a5 = np.hstack((ones, SIG(z5)))
    
    z6 = a5 @ initt_wt5.T
    
    a6 = SIG(z6)
    print(a6.shape)
    
    
    d_6 = a6-y
    
    
    l = len(Train_otpts)

    ones = np.ones((l,1))
    
    z5=np.hstack((ones, z5))
    d_5=np.multiply(initt_wt5.T @ d_6.T, SIGGrad(z5).T)
    
    z4 = np.hstack((ones, z4))
    d_4 = np.multiply(initt_wt4.T @ d_5[1:], SIGGrad(z4).T)
    
    z3 = np.hstack((ones, z3))
    d_3 = np.multiply(initt_wt3.T @ d_4[1:], SIGGrad(z3).T)
    z2 = np.hstack((ones, z2))
    d_2 = np.multiply(initt_wt2.T @ d_3[1:], SIGGrad(z2).T)
    
    Delta1 = np.zeros(initt_wt1.shape)
    Delta2 = np.zeros(initt_wt2.shape)
    Delta3 = np.zeros(initt_wt3.shape)
    Delta4 = np.zeros(initt_wt4.shape)
    Delta5 = np.zeros(initt_wt5.shape)

   
    Delta1 = Delta1 + d_2[1:,:] @ a1
    print('Delta1',Delta1.shape)
    Delta2 = Delta2 + d_3[1:,:] @ a2
    print('Delta2',Delta2.shape)
    Delta3 = Delta3 + d_4[1:,:] @ a3
    print('Delta3',Delta3.shape)
    Delta4 = Delta4 + d_5[1:,:] @ a4
    print('Delta4',Delta4.shape)
    Delta5 = Delta5 + d_6.T @ a5



    Delta1 /= l
    Delta2 /= l
    Delta3 /= l
    Delta4 /= l
    Delta5 /= l

     
     
        
    return np.hstack((Delta1.ravel(order='F'), Delta2.ravel(order='F'),Delta3.ravel(order='F'),Delta4.ravel(order='F'),Delta5.ravel(order='F'))) 
```

```python
bckprop_wts = NN(ttl_weits, ip_lyr, hd_lyr1,hd_lyr2,hd_lyr3,hd_lyr4, no_labls,Train_ip,Train_otpts)

```

```python
def predicting_y(wt1, wt2, wt3,wt4,wt5,X, y):
    l = len(X)
   
    ones = np.ones((l,1))
    
    a1 = np.hstack((ones, y))
    
    a2 = SIG(a1 @ wt1.T)
    
    a2 = np.hstack((ones, a2))
    
    a3 = SIG(a2 @ wt2.T)
    
    a3 = np.hstack((ones, a3))
    
    a4 = SIG(a3 @ wt3.T)
    
    a4 = np.hstack((ones, a4))
    
    a5 = SIG(a4 @ wt4.T)
    
    a5 = np.hstack((ones, a5))
    
    h = SIG(a5 @ wt5.T)
    return  np.argmax(h, axis = 1)
```

```python

def Gard_checking(ttl_weits,bckprop_wts,ip_lyr, hd_lyr1,hd_lyr2,hd_lyr3,hd_lyr4, no_labls,X,y):
    eps = 0.22 
    l = len(ttl_weits) 
    
    for i in range(5):
        iterattion = int(np.random.rand()*len(ttl_weits))
        
        adding_eps = np.zeros((len(ttl_weits),1))
        
        adding_eps[iterattion] = eps
        ThetaPlus = Feedfrwd(ttl_weits + adding_eps.flatten(),ip_lyr, hd_lyr1,hd_lyr2,hd_lyr3,hd_lyr4, no_labls,X,y)
        
        ThetaMinus = Feedfrwd(ttl_weits - adding_eps.flatten(),ip_lyr, hd_lyr1,hd_lyr2,hd_lyr3,hd_lyr4, no_labls,X,y)
        
        
        num_app = (ThetaPlus - ThetaMinus) / float(2*eps)
        print('num_app',num_app)
       
        print('bckprop_wts[iterattion]',bckprop_wts[iterattion])
        
        
        
```

```python
grad_chk=Gard_checking(ttl_weits,bckprop_wts,ip_lyr, hd_lyr1,hd_lyr2,hd_lyr3,hd_lyr4, no_labls,Train_ip,Train_otpts)


grad_chk
```

```python
import scipy.optimize as opt
Final_weights = opt.fmin_cg(maxiter = 100, f = Feedfrwd, x0 = ttl_weits, fprime = NN,
                        args = (ip_lyr, hd_lyr1,hd_lyr2,hd_lyr3, hd_lyr4,no_labls,Train_ip,Train_otpts))

Final_weights



```

```python

Final_wts1,Final_wts2,Final_wts3,Final_wts4,Final_wts5 = Weights(Final_weights,ip_lyr, hd_lyr1,hd_lyr2,hd_lyr3,hd_lyr4, no_labls)

    
```

## Predicting y values for test_data

```python
pred = predicting_y(Final_wts1,Final_wts2,Final_wts3,Final_wts4,Final_wts5,X_test.iloc[:,:-1], y_test)
pred
```

```python
y_actual=X_test.iloc[:,-1]
```
